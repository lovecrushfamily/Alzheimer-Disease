```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from mealpy import BinaryVar, GA, Problem
import warnings
from sklearn.exceptions import ConvergenceWarning
import matplotlib.pyplot as plt
from IPython.display import clear_output

warnings.filterwarnings('ignore', category=ConvergenceWarning)

fitness_histories = {}
fig, ax = plt.subplots(figsize=(10, 6))

class FeatureSelectionProblem(Problem):
    def __init__(self, bounds=None, minmax="max", X_train=None, y_train=None, classifier=None, classifier_name=None, **kwargs):
        self.X_train = X_train
        self.y_train = y_train
        self.classifier = classifier
        self.classifier_name = classifier_name
        self.fitness_history = []
        super().__init__(bounds, minmax, **kwargs)

    def obj_func(self, x):
        selected_features = x.astype(bool)
        
        if not selected_features.any():
            return 0.0
        
        X_selected = self.X_train[:, selected_features]
        
        if callable(self.classifier):
            model = self.classifier()
        else:
            model = self.classifier.__class__(**self.classifier.get_params())
            
        model.fit(X_selected, self.y_train)
        y_val_pred = model.predict(X_selected)
        fitness_score = accuracy_score(self.y_train, y_val_pred)
        
        
        self.fitness_history.append(fitness_score)
        
        
        self.update_plot()
        
        return fitness_score
    
    def update_plot(self):
        clear_output(wait=True)
        ax.clear()
        
        for name, history in fitness_histories.items():
            if len(history) > 0:  
                ax.plot(history, label=f'{name}')
        
        if len(self.fitness_history) > 0:
            if self.classifier_name not in fitness_histories:
                fitness_histories[self.classifier_name] = []
            fitness_histories[self.classifier_name] = self.fitness_history
            ax.plot(self.fitness_history, label=f'{self.classifier_name} (Current)')
        
        ax.set_xlabel('Evaluations')
        ax.set_ylabel('Fitness Score')
        ax.set_title('Genetic Algorithm Progress')
        ax.legend()
        ax.grid(True)
        plt.tight_layout()
        plt.show()

term_dict = {
    "max_epoch": 25,
    "max_time": 60, 
    "max_early_stop": 10
}

classifiers = {
    "ANN": lambda: MLPClassifier(max_iter=1000, learning_rate_init=0.001, solver='adam'),
    "SVM": SVC,
    "Random Forest": RandomForestClassifier,
}

bounds = BinaryVar(n_vars=X.shape[1], name="feature_selection")
results = {}

for name, clf in classifiers.items():
    print(f"\nOptimizing {name}...")
    problem = FeatureSelectionProblem(
        bounds=bounds, 
        minmax="max", 
        X_train=X_train_std, 
        y_train=y_train, 
        classifier=clf,
        classifier_name=name
    )
    model = GA.BaseGA(epoch=25, pop_size=50, pc=0.9, pm=0.05)
    
    best_solution = model.solve(problem, termination=term_dict)
    
    results[name] = {
        "Best feature selection": best_solution.solution,
        "Best fitness score": best_solution.target.fitness,
        "Fitness history": problem.fitness_history
    }

print("\nFinal Results:")
for name, result in results.items():
    print(f"\n{name}:")
    print(f"  Best fitness score: {result['Best fitness score']:.4f}")
    
    
    feature_names = df.drop(columns='Diagnosis').columns
    selected_features = feature_names[result['Best feature selection'].astype(bool)]
    print(f"  Selected features ({len(selected_features)} features):")
    for feature in selected_features:
        print(f"    - {feature}")


plt.figure(figsize=(12, 6))
for name, result in results.items():
    plt.plot(result['Fitness history'], label=name)
plt.xlabel('Evaluations')
plt.ylabel('Fitness Score')
plt.title('Genetic Algorithm Progress Comparison')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
